```python
import pandas as pd
import os 
import sys
cwd = os.getcwd()

pd.set_option('display.max_rows', 30)
pd.set_option('display.max_columns', None)
```

```python
import os

fpl_path = os.path.join(cwd, '..', '..', 'data', 'history', 'merged_gw_2324.csv')
team_path = os.path.join(cwd, '..', '..', 'data', 'history', 'teams_2324.csv')
fixture_path= os.path.join(cwd, '..', '..', 'data', 'history', 'fixtures_2324.csv')
data_folder=os.path.join(cwd, '..', '..', 'data', 'history')
df_player = pd.read_csv(fpl_path)
df_team=pd.read_csv(team_path)
df_fixture=pd.read_csv(fixture_path)
```

```python
df_player
```

```python
df_player = df_player.rename(columns={
    'name': 'Player',
    'position': 'Pos',
    'team': 'Team_player',
    'xP': 'xP',
    'assists': 'A',
    'bonus': 'B',
    'bps': 'BPS',
    'clean_sheets': 'CS',
    'creativity': 'C',
    'element': 'element',
    'expected_assists': 'xA',
    'expected_goal_involvements': 'xGI',
    'expected_goals': 'xG',
    'expected_goals_conceded': 'xGC',
    'fixture': 'fixture',
    'goals_conceded': 'GC',
    'goals_scored': 'GS',
    'ict_index': 'ICT',
    'influence': 'I',
    'kickoff_time': 'kickoff_time',
    'minutes': 'Mins',
    'opponent_team': 'vs',
    'own_goals': 'OG',
    'penalties_missed': 'Pen_Miss',
    'penalties_saved': 'Pen_Save',
    'red_cards': 'RC',
    'round': 'round',
    'saves': 'S',
    'selected': 'SB',
    'starts': 'starts',
    'team_a_score': 'team_a_score',
    'team_h_score': 'team_h_score',
    'threat': 'T',
    'total_points': 'Pts',
    'transfers_balance': 'transfers_balance',
    'transfers_in': 'Tran_In',
    'transfers_out': 'Tran_Out',
    'value': 'Price',
    'was_home': 'was_home',
    'yellow_cards': 'YC'
})
```

```python
df_player['season']=2324
```

```python
df_player = df_player.drop(['xP', 'element', 'fixture', 'starts', 'transfers_balance','round'], axis=1)
```

```python
df_player
```

```python
df_player = df_player.merge(df_team[['id', 'name']], how='left', left_on='Team_player', right_on='name')
```

```python
df_player
```

```python
# Merge for home games
merged_home = pd.merge(
    df_player[df_player['was_home'] == True],  # Filter for home games
    df_fixture[['event', 'team_h', 'team_a', 'team_h_difficulty', 'team_a_difficulty']],
    left_on=['id', 'vs'],  # 'team' is the player's team, 'vs' is the opponent team
    right_on=['team_h', 'team_a'],  # Join home team ('team_h') and away team ('team_a')
    how='left'
)

# Merge for away games
merged_away = pd.merge(
    df_player[df_player['was_home'] == False],  # Filter for away games
    df_fixture[['event', 'team_h', 'team_a', 'team_h_difficulty', 'team_a_difficulty']],
    left_on=['id', 'vs'],  # 'team' is the player's team, 'vs' is the opponent team
    right_on=['team_a', 'team_h'],  # Join away team ('team_a') and home team ('team_h')
    how='left'
)
# Combine both home and away merges
final_merged = pd.concat([merged_home, merged_away], ignore_index=True)
```

```python
final_merged
```

```python
final_merged['Team_fdr'] = final_merged.apply(
    lambda row: row['team_h_difficulty'] if row['was_home'] else row['team_a_difficulty'], axis=1
)
final_merged['opponent_fdr'] = final_merged.apply(
    lambda row: row['team_a_difficulty'] if row['was_home'] else row['team_h_difficulty'], axis=1
)

# Step 3: Drop the columns from df_fixture that are no longer needed
merged_fdr = final_merged.drop(columns=['team_a', 'team_h_difficulty', 'team_a_difficulty','id','name','event','team_h'])
```

```python
merged_fdr
```

```python
for index, row in merged_fdr.iterrows():
    if row['was_home'] == True:
        # If was_home is True, swap team_a_score and team_h_score
        merged_fdr.at[index, 'team_player_score'] = row['team_h_score']
        merged_fdr.at[index, 'vs_score'] = row['team_a_score']
    else:
        # If was_home is False, swap team_a_score and team_h_score
        merged_fdr.at[index, 'vs_score'] = row['team_h_score']
        merged_fdr.at[index, 'team_player_score'] = row['team_a_score']

# Drop the original 'team_h_score' and 'team_a_score' columns
merged_fdr = merged_fdr.drop(columns=['team_h_score', 'team_a_score'])
```

```python
merged_fdr['Price'] = merged_fdr['Price'].astype(float)/10
```

```python
merged_fdr
```

```python
merged_fdr = merged_fdr.merge(df_team[['short_name', 'name']], how='left', left_on='Team_player', right_on='name')

# Replace the 'Team_player' column with 'short_name'
merged_fdr['Team_player'] = merged_fdr['short_name']

# Drop the 'short_name' column after replacing
merged_fdr = merged_fdr.drop(columns=['short_name', 'name'])

```

```python
merged_fdr = merged_fdr.merge(df_team[['short_name', 'id']], how='left', left_on='vs', right_on='id')

# Replace the 'vs' column with 'short_name'
merged_fdr['vs'] = merged_fdr['short_name']

# Drop the 'short_name' and 'name' columns after replacing
merged_fdr = merged_fdr.drop(columns=['short_name', 'id'])
```

```python
def convert_opponent_string(df):
    df.loc[df['was_home'] == True, 'vs'] = df['vs'] + ' (A)'
    df.loc[df['was_home'] == False, 'vs'] = df['vs'] + ' (H)'
    df.loc[df['was_home'] == True, 'Team_player'] = df['Team_player'] + ' (H)'
    df.loc[df['was_home'] == False, 'Team_player'] = df['Team_player'] + ' (A)'
    return df
```

```python
merged_home = pd.merge(merged_fdr, df_team[['short_name', 
                                                    'strength_overall_home', 
                                                    'strength_overall_away', 
                                                    'strength_attack_home', 
                                                    'strength_attack_away', 
                                                    'strength_defence_home', 
                                                    'strength_defence_away']],
                       left_on='Team_player', 
                       right_on='short_name', 
                       how='left')

# Merge for the opponent team
merged_opponent = pd.merge(merged_home, 
                            df_team[['short_name', 
                                       'strength_overall_home', 
                                       'strength_overall_away', 
                                       'strength_attack_home', 
                                       'strength_attack_away', 
                                       'strength_defence_home', 
                                       'strength_defence_away']],
                            left_on='vs', 
                            right_on='short_name', 
                            how='left', 
                            suffixes=('', '_opponent'))

# Optionally drop the 'short_name' columns for opponents if you don't need them
merged_opponent = merged_opponent.drop(columns=['short_name', 'short_name_opponent'])
merged_opponent=convert_opponent_string(merged_opponent)
```
